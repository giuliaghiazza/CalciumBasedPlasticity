# Now we can use the new module

## Pre-before-Post

```python
nest.ResetKernel()
nest.Install("PW_module")
duration = 10000 # ms
frequency = 10.0  # Hz
delta_t = 10.0  # ms

neuron_model = "mhill_tononi_neuron_nestml__with_stdp_nestml"
syn_model = "stdp_nestml__with_mhill_tononi_neuron_nestml"

pre_neuron = nest.Create(neuron_model, 1)
post_neuron = nest.Create(neuron_model, 1)


nest.SetStatus(pre_neuron, {"theta_d": 0.6, "theta_p": 0.74})
nest.SetStatus(post_neuron, {"theta_d": 0.6, "theta_p": 0.74})
nest.SetStatus(pre_neuron, {"gamma_p_GB": 216.2})
nest.SetStatus(post_neuron, {"gamma_p_GB": 216.2})
nest.SetStatus(pre_neuron, {"gamma_d_GB": 101.5})
nest.SetStatus(post_neuron, {"gamma_d_GB": 101.5})
nest.SetStatus(pre_neuron, {"gmax_d_AMPA": 0.5})
nest.SetStatus(post_neuron, {"gmax_p_AMPA": 1.0})
nest.SetStatus(pre_neuron, {"gmax_d_AMPA": 0.5})
nest.SetStatus(post_neuron, {"gmax_p_AMPA": 1.0})


wr = nest.Create("weight_recorder")
nest.SetDefaults(syn_model, {"weight_recorder": wr[0]})
receptors = nest.GetDefaults("mhill_tononi_neuron_nestml__with_stdp_nestml")["receptor_types"]
syn_spec = nest.CollocatedSynapses(
                                    {'synapse_model':syn_model,
                                     'receptor_type':receptors["AMPA"],
                                     }
                                    ,
                                   {
                                    'synapse_model':syn_model,
                                    'receptor_type': receptors["NMDA"],
                                       }
                                   )
nest.Connect(pre_neuron, post_neuron, "one_to_one", syn_spec=syn_spec)

synapse = nest.GetConnections(pre_neuron, post_neuron)
nest.SetStatus(synapse, {"Wmax":50.0})

interval = 1000.0 / frequency
spike_times = np.arange(interval, duration, interval)
spike_generator = nest.Create("spike_generator", {"spike_times": spike_times})

if delta_t >= 0:
    delta_pre = 0.0
    delta_post = delta_t

else:
    delta_pre = -1 * delta_t
    delta_post = 0.0

    
syn_spec_spike_pre = nest.CollocatedSynapses(
                                    {'synapse_model':'static_synapse',
                                     'receptor_type': 3,
                                     'weight': 999.0,
                                     'delay': 1.0 + delta_pre,
                                    }
                                   )
syn_spec_spike_post = nest.CollocatedSynapses(
                                    {'synapse_model':'static_synapse',
                                     'receptor_type': 3,
                                     'weight': 999.0,
                                     'delay': 1.0 + delta_post,
                                    }
                                   )
    
nest.Connect(spike_generator, pre_neuron, "one_to_one", syn_spec=syn_spec_spike_pre)
nest.Connect(spike_generator, post_neuron, "one_to_one", syn_spec=syn_spec_spike_post)

sd_1 = nest.Create("spike_recorder")
sd_2 = nest.Create("spike_recorder")
nest.Connect(pre_neuron, sd_1)
nest.Connect(post_neuron, sd_2)

vm_pre = nest.Create("voltmeter")
vm_post = nest.Create("voltmeter")
nest.Connect(vm_pre, pre_neuron)
nest.Connect(vm_post, post_neuron)

recorder = nest.Create( "multimeter",params={"interval": delta_t, "record_from": ["cai_CR", "effcai_GB", "rho_GB", "I_syn_vdcc", "I_syn_nmda", "AMPA_g_peak", "I_syn_ampa"]})
nest.Connect(recorder, post_neuron)

nest.Simulate(duration)

spikes_pre = sd_1.get(['events'][0])
spikes_post = sd_2.get(['events'][0])
#print(sd.events)
```

```python
ev       = recorder.get("events")
rho_GB      = ev["rho_GB"]
I_syn_vdcc = ev["I_syn_vdcc"]
cai_CR = ev["cai_CR"]
effcai_GB = ev["effcai_GB"]

max_rho = max(rho_GB)
max_I_vdcc = max(I_syn_vdcc)
max_cai_CR = max(cai_CR)
max_effcai_GB = max(effcai_GB)

print("Massimo valore di rho_GB:", max_rho)
print("Massimo valore di I_syn_vdcc:", max_I_vdcc)
print("Massimo valore di cai_CR:", max_cai_CR)
print("Massimo valore di effcai_GB:", max_effcai_GB)
```

```python
ev       = recorder.get("events")
t        = ev["times"]
cai_CR      = ev["cai_CR"]
effcai_GB = ev["effcai_GB"]
rho_GB      = ev["rho_GB"]
I_syn_vdcc  = ev["I_syn_vdcc"]
I_syn_nmda = ev["I_syn_nmda"]
AMPA_g_peak = ev["AMPA_g_peak"]


fig, ax = plt.subplots(9, figsize = (18,20))

ax[0].plot(vm_pre.events["times"], vm_pre.events["V_m"], color = "b", label = "Pre-synaptic potential")
ax[0].plot(vm_post.events["times"], vm_post.events["V_m"], color = "orange",  label = "Post-synaptic potential")
ax[0].set_title("Post-synaptic neuron Potential")
ax[0].set_ylabel("Vm [mV]")
ax[0].legend()

ax[1].plot(vm_pre.events["times"], vm_pre.events["V_m"], color = "b", label = "Pre-synaptic potential")
ax[1].plot(vm_post.events["times"], vm_post.events["V_m"], color = "orange",  label = "Post-synaptic potential")
ax[1].set_title("Post-synaptic neuron Potential")
ax[1].set_ylabel("Vm [mV]")
ax[1].set_xlim([7700,9500])
ax[1].legend()

ax[2].plot(wr.events["times"], wr.events["weights"], color = 'r')
ax[2].set_title("STDP")
ax[2].set_ylabel("Weight [nS]")

ax[3].plot(recorder.events["times"], recorder.events["cai_CR"], color = "b", label = "Concentrazione interna calcio post")
ax[3].set_title("Concentrazione interna calcio post")
ax[3].set_ylabel("cai_CR [nmol/(dm**3)]")
ax[3].legend()

ax[4].plot(recorder.events["times"], recorder.events["effcai_GB"], color = "g", label = "effcai_GB")
ax[4].set_title("effcai_GB")
ax[4].set_ylabel("effcai_GB")
ax[4].legend()

ax[5].plot(recorder.events["times"], recorder.events["rho_GB"], color = "orange", label = "rho")
ax[5].set_title("rho")
ax[5].set_ylabel("rho_GB []")
ax[5].legend()

ax[6].plot(recorder.events["times"], recorder.events["I_syn_vdcc"], color = "r", label = "I_VDCC")
ax[6].set_title("I_VDCC")
ax[6].set_ylabel("I_syn_vdcc [pA]")
ax[6].legend()
ax[6].set_xlabel("Time [ms]")

ax[7].plot(recorder.events["times"], recorder.events["I_syn_nmda"], color = "b", label = "I_NMDA")
ax[7].set_title("I_NMDA")
ax[7].set_ylabel("I_syn_nmda [pA]")
ax[7].legend()
ax[7].set_xlabel("Time [ms]")

ax[8].plot(recorder.events["times"], recorder.events["AMPA_g_peak"], color = "g", label = "AMPA_g_peak")
ax[8].set_title("AMPA_g_peak")
ax[8].set_ylabel("AMPA_g_peak [nS]")
ax[8].legend()
ax[8].set_xlabel("Time [ms]")

fig.tight_layout()
```

```python
plt.figure(figsize = (18,10))
plt.plot(recorder.events["times"], recorder.events["effcai_GB"])
plt.axhline(y=0.74, linestyle='--', color='orange', label='theta_p')
plt.axhline(y=0.6, linestyle='--', color='green', label='theta_d')
plt.ylabel("effcai_GB")
plt.xlabel("t [ms]")
plt.title("effcai_GB - Pre before Post")
plt.legend()
plt.show()
```
